#### Data visualizations and Exploratory Data Analysis

```python
#let's see what hosts (IDs) have the most listings on Airbnb platform and taking advantage of this service
top_host=df.host_id.value_counts().head(10)
top_host
```

```python
#we can confirm our fidnings with already existing column called 'calculated_host_listings_count'
top_host_check=df.calculated_host_listings_count.max()
top_host_check
```

```python
top_host_df=pd.DataFrame(top_host)
top_host_df.reset_index(inplace=True)
top_host_df.rename(columns={'index':'Host_ID', 'host_id':'P_Count'}, inplace=True)
top_host_df
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Host_ID", y="P_Count", data=top_host_df,palette='Blues_d')
plt.title('Hosts with the most listings in NYC')
plt.ylabel('Count of listings')
plt.xlabel('Host IDs')
```

we can see that there is a good distribution between top 10 hosts with the most listings. First host has more than 300+ listings

```python
#host level analysis
host_stats=df.groupby('host_id').agg({'price':'mean','minimum_nights':'mean','number_of_reviews':'mean'}).reset_index()
print(host_stats.describe())
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=host_stats,x='number_of_reviews',y='price')
plt.title('Total Reviews vs. Reviews per month by host')
plt.xlabel('Total Reviews')
plt.ylabel('Average price')
plt.show()
```

A scatter plot is created to view total reviews by average price using host statistics

#### Barchart for categorical data : the distribution of listing in different areas

```python
plt.figure(figsize=(10,6))
sns.countplot(x='neighbourhood_group',data=df,palette='viridis')
plt.title("Distribution of listing in neighbourhood group")
plt.xlabel("neighbourhood_group")
plt.ylabel("count")
plt.show()
```

we can state that Manhattan has the highest no of counts. Staten Island appear to have less count.

```python
#for the sake of a better visualization,extreme values are avoided
plt.figure(figsize=(10,6))
#creating a sub-dataframe with no extreme values / less than 500
sub_df=df[df.price < 500]
#using violinplot to showcase density and distribtuion of prices 
sns.boxplot(data=sub_df, x='neighbourhood_group', y='price')
plt.title('Density and distribution of prices for each neighberhood_group')
plt.xlabel("neighbourhood_group")
plt.ylabel("price")
plt.show()
```

Great, with box plot we can definitely observe a couple of things about distribution of prices for Airbnb in NYC boroughs. First, we can state that Manhattan has the highest range of prices for the listings. Queens and Staten Island appear to have very similar distributions, Bronx is the cheapest of them all. This distribution and density of prices were completely expected; for example, as it is no secret that Manhattan is one of the most expensive places in the world to live in, where Bronx on other hand appears to have lower standards of living.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='longitude',y='latitude',data=df, c='blue',alpha=0.2)
plt.title("Airbnb Listings In New York City")
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.show()
```

Good, scatterplot worked just fine to output our latitude and longitude points.

#### count plot to visualise the count of roomtype

```python
plt.figure(figsize=(8,6))
sns.countplot(x='room_type',data=df,palette='Set3')
plt.title("Distribution of Roomtypes")
plt.xlabel("room_type")
plt.ylabel("count")
plt.show()
```

#### Histograms for numerical data : price,minimum nights,availability 365

```python
fig,axes=plt.subplots(1,3,figsize=(15,5))
axes[0].hist(df['price'],bins=30,color='skyblue')
axes[0].set_title('Price Distribution')
axes[0].set_xlabel('Price')
axes[0].set_ylabel('Frequency')
axes[1].hist(df['minimum_nights'],bins=30,color='salmon')
axes[1].set_title('Minimun Nights Distribution')
axes[1].set_xlabel('Minimum Nights')
axes[1].set_ylabel('Frequency')
axes[2].hist(df['availability_365'],bins=30,color='lightgreen')
axes[2].set_title('Availability 365 Distribution')
axes[2].set_xlabel('Availability 365')
axes[2].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Here, shows the frequency distribution of price,minimum nights and availability 365.
